```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/MyDrive/디영처/lena2.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gImg= cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg,cmap='gray')
```

```python
outImg = cv2.GaussianBlur(gImg, (3,3), 1)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (3,3), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7,7), 1)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7,7), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (15,15), 3)
cv2_imshow(outImg)
```

```python
mean_img = cv2.blur(gImg, (5, 5))
cv2_imshow(mean_img)
```

```python
edge_img = cv2.addWeighted(gImg, 1.0, mean_img, -1.0, 0)
cv2_imshow(edge_img)
```

```python
out_img = cv2.addWeighted(gImg, 1.0, edge_img, 1.0, 0)
cv2_imshow(out_img)
```
